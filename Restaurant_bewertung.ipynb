```python




js_reviews = []
for review in restaurantreviews:
    #print("============= REVIEW ===================")
    js_review = []
    
    reviewspans= review.find_all() 
    for reviewspan in reviewspans:
        js_review_part = {}
        if 'class' in reviewspan.attrs: 
            js_review_part['class'] = reviewspan.attrs['class']
        if 'itemprop' in reviewspan.attrs: 
            js_review_part['itemprop'] = reviewspan.attrs['itemprop'] 
            
        if 'style' in reviewspan.attrs: 
            js_review_part['style'] =  reviewspan.attrs['style']     
        if (len(reviewspan.findChildren())) == 0:
            js_review_part['text'] =  reviewspan.contents
        js_review.append(dict(js_review_part))
    js_reviews.append(js_review)
```
